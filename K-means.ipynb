```python
df=pd.read_csv("iris.data",header=None)
df.iloc[:,4]=df.iloc[:,4].map({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
df
```

```python
d1=df.iloc[:,:2]
d1
```

```python
k=3   #Number of clusters
n=d1.shape[0]
c=d1.shape[1]
```

```python
cr=d1.sample(3)
d1=np.array(d1,dtype=float)
cr=np.array(cr,dtype=float)
```

```python
def distance(cr,d1):
    a=np.empty([d1.shape[0],k])
    for i in range(cr.shape[0]):
        for j in range(d1.shape[0]):
            d=np.sum(np.square(cr[i]-d1[j]))
            #print(d)
            #a=np.full((150,2),d)
            a[j][i]=d
    return a
```

```python
dist = distance(cr,d1)
dist
```

```python
def assignCluster(dist):
    #mi=np.min(dist,axis=1)
    #print(mi)
    c1 = []
    c2 = []
    c3 = []
    minindx= np.argmin(dist,axis=1)
    for i in range(len(d1)):
        if minindx[i]==0:
            c1.append(d1[i])
        elif minindx[i]==1:
            c2.append(d1[i])
        else:
            c3.append(d1[i])
    return (c1,c2,c3)
    
    
```

```python
c1,c2,c3 = assignCluster(dist)
c3
```

```python
def updateCR(cr):
    cr = np.zeros(cr.shape)
    c1,c2,c3 = assignCluster(dist)
    for i in range(len(c1)):
        cr[0][0] += c1[i][0]
        cr[0][0] = cr[0][0]/len(c1)
        cr[0][1] += c1[i][1]
        cr[0][1] = cr[0][1]/len(c1)
    for i in range(len(c2)):
        cr[1][0] += c2[i][0]
        cr[1][0] = cr[1][0]/len(c2)
        cr[1][1] += c2[i][1]
        cr[1][1] = cr[1][1]/len(c2)
    for i in range(len(c3)):
        cr[2][0] += c3[i][0]
        cr[2][0] = cr[2][0]/len(c3)
        cr[2][1] += c3[i][1]
        cr[2][1] = cr[2][1]/len(c3)
    return cr
```

```python
def KMeans(thres,cr,d1):
    for i in range(thres):
        dist = distance(cr,d1)
        #c1,c2,c3 = assignCluster(dist)
        cr = updateCR(cr)
    return (c1,c2,c3)
```
